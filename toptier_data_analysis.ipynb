```python
# Replace authors countries with authors from affiliation history in case of Africans only
import numpy as np

fnumbers = [0, 15675, 22127, 3315, 39574]
aff_history = {}
for fno in fnumbers:
  aff_history.update(np.load("/content/drive/MyDrive/Ro'ya CV4Africa Community Files/Bibliometric Study/Work_Phases/Data_Collection/toptier_authors_affiliation_history/authors_aff_history_%d.npy"%fno, allow_pickle=True).item())
  print(fno, ' : ', len(aff_history))

print(len(aff_history))
#authors.iloc[authors.auid==6506184188, 9] = 'United States'
#display(authors)
print(authors[authors.surname=='Siam']['auid'])

def retrieve_africa_from_history(auid, original_country):
  if auid not in aff_history:
    non_existant_authors.append(original_country)
    return original_country

  history = aff_history[auid]

  for country in history:
    if country is None:
      continue

    if country.lower() in african_countries:
      return country

  return original_country

def check_affiliation_history(authorss):
  for idx, author in authorss.iterrows():
    if author['country'] in african_countries:
      continue

    new_country = retrieve_africa_from_history(author['auid'], author['country'])
    if new_country != author['country']:
      print(author['surname'], ' ', author['given_name'], ' ', author['eid'], ' : ', author['country'], ' ', new_country)

    authorss.iloc[authorss.auid==author.auid, 9] = new_country
  return authorss

def check_confirmed_african_authors(authorss, global_counter):
  for idx, author in authorss.iterrows():
    if author['country'] in african_countries:
      continue

    new_country = author['country']
    for entry in african_authors_nonafrican_inst:
      if author['surname'] == entry[0] and author['given_name'] == entry[1]:
        new_country = entry[-1]
        global_counter += 1

    if new_country != author['country']:
      print(author['surname'], ' ', author['given_name'], ' ', author['eid'], ' : ', author['country'], ' ', new_country)

    authorss.iloc[authorss.auid==author.auid, 9] = new_country
  return authorss, global_counter
#print(authors[authors.eid=='2-s2.0-85078804967']['country'].iloc[0])
#display(check_affiliation_history(authors[authors.auid==55660049000]))
```

```python
african_authors_nonafrican_inst = []
with open("/content/drive/MyDrive/Ro'ya CV4Africa Community Files/Bibliometric Study/Work_Phases/Data_Analysis/confirmed_africans.txt", 'r') as f:
  for line in f:
    tokens = line.split(" ")

    surname = tokens[0].strip()
    if surname == 'Sabry':
      given_name = 'Aly   Mohamed'
    elif "Baji" in surname:
      given_name = 'Ivan V.'
    else:
      given_name = tokens[3].strip()

    country = tokens[-1].strip()
    if country == 'Africa':
      country = 'South Africa'

    african_authors_nonafrican_inst.append( (surname, given_name, country))
    print(given_name, ' ', surname, ' : ', country)

print(len(african_authors_nonafrican_inst))
```

## Preprocessing intermediate csvs

```python
# Create a dataframe with number of publications per year per country
# Columns are sorted based on the years

non_existant_authors = []

filtered_list = ['VGRAPH: An effective approach for generating static video summaries',
                 'The visual object tracking VOT2013 challenge results',
                 'Enhanced target tracking in UAV imagery with P-N learning and structural constraints',
                 'Robust real-time tracking with diverse ensembles and random projections',
                 'Residual Dilated U-net for the Segmentation of COVID-19 Infection from CT Images',
                 'Multiview active shape models with SIFT descriptors for the 300-W face landmark challenge']

def check_filtered_pub(pub_title):
  if pub_title in filtered_list:
    return True
  else:
    return False

global_counter = 0

def check_cv_confs(venue):
  cv_confs = ["Proceedings of the IEEE Computer Society Conference on Computer Vision and Pattern Recognition",
             "IEEE Conference on Computer Vision and Pattern Recognition"]

  for cv_conf in cv_confs:
    if cv_conf in venue:
      return True
  return False

results = pd.DataFrame()
african_pubs = {}

#### TODO: use filter on top tier venues
venues_filter = []

cvonly = False
aff_history_into_account = False

keys = []
for index, row in pubs.iterrows():
    if row['Year'] not in results.columns:
        results[row['Year']] = 0

    if row['subtypeDescription'] in ['Retracted', 'Editorial', 'Review', 'Erratum', 'Conference Review']:
      # Wrongly counted part of publications
      continue

    if check_filtered_pub(row['title']):
      continue

    if cvonly:
      if not check_cv_confs(row['publicationName']):
        continue

    temp_authors = authors[authors['eid'] == row['eid']]
    temp_authors = temp_authors.drop(temp_authors[temp_authors.country == 'Swaziland'].index)
    #To not count the same author in the same country twice
    temp_authors = temp_authors.drop_duplicates(subset=['country', 'indexed_name', 'surname', 'given_name'])

    if aff_history_into_account:
      temp_authors = temp_authors[temp_authors[['country']].notnull().all(1)]
      temp_authors, global_counter = check_confirmed_african_authors(temp_authors, global_counter)
      #display(temp_authors['country'])

    temp_authors = temp_authors['country'].value_counts()
    for key, value in temp_authors.items():
        key = key.lower()
        if key not in results.index:
            results.loc[key] = 0
            keys +=  [key]

        if key in african_countries:
          if key not in african_pubs:
            african_pubs[key] = []
          african_pubs[key].append(row)

        results.loc[key, row['Year']] = results.loc[key, row['Year']] + value
results = results.sort_index(axis=1, ascending=False).reset_index().rename(columns={'index': 'Country'})

if cvonly:
  outfile = 'pubs_by_country_cvonly.csv'
else:
  outfile = 'pubs_by_country.csv'
results.to_csv(outfile)
#np.save('non_existant_authors.npy', non_existant_authors)
#print('Number of found africans in nonAfrican inst ', global_counter)
```

## Generating the Bar Chart

```python
# Generating the bar chart data by counting the number of publications per region per year
# All years before 2012 are summed togther

#results = pd.read_csv('pubs_by_country.csv')

assert len(results) == len(keys)

african_pubs_count = {}

total_publications_per_year = {}
total_african_publications_per_year = {}

for i in range(len(results)):
  for year in range(2012, 2023):
    if year not in total_publications_per_year:
      total_publications_per_year[year] = 0
      total_african_publications_per_year[year] = 0

    if results.loc[i, 'Country'] in african_countries:
      total_african_publications_per_year[year] += results.loc[i, year]

      if results.loc[i, 'Country'] not in african_pubs_count:
        african_pubs_count[results.loc[i, 'Country']] = 0
      african_pubs_count[results.loc[i, 'Country']] += results.loc[i, year]

    total_publications_per_year[year] += results.loc[i, year]

for k, v in total_publications_per_year.items():
  print('Year ', k, ' : ', total_african_publications_per_year[k]/total_publications_per_year[k] * 100, ' ', total_african_publications_per_year[k])

african_pubs_count = sorted(african_pubs_count.items(), key=lambda x:x[1])
for country, rate in african_pubs_count:
  print('African country: ', country, ' : ', rate)
  for row in african_pubs[country]:
    print(row['title'])

import numpy as np
print('Mean', np.mean([v/total_publications_per_year[k] for k, v in total_african_publications_per_year.items()])*100)
```

```python
def retrieve_continent(country, continents):
  for cont, countries in continents.items():
    if country in countries:
      return cont

  raise Exception('Country not found ', country)

continents = {'North America': [n.lower() for n in my_world.get_countries_list_of('North America')],
              'South America': [n.lower() for n in my_world.get_countries_list_of('South America')],
              'Africa': african_countries,
              'Asia': [n.lower() for n in my_world.get_countries_list_of('Asia')] + \
                    ['russian federation', 'viet nam', 'macao'],
              'Europe': [n.lower() for n in my_world.get_countries_list_of('Europe')],
              'Oceania': [n.lower() for n in my_world.get_countries_list_of('Oceania')] + ['martinique']}



continents_list = []
for idx, row in results.iterrows():
    continents_list.append(retrieve_continent(row['Country'], continents))
results['Continent'] = continents_list

display(results)
```

```python


results_final = pd.DataFrame()
for cont in continents.keys():
    results_final[cont] = results[results['Continent']==cont].iloc[:, 1:-1].sum(axis=0)

if cvonly:
  results_final.to_csv('toptier_cvonly_plot.csv')
else:
  results_final.to_csv('toptier_plot.csv')
```
